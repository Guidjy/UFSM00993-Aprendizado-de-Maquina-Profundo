```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# gera 20 pontos aleatórios entre 0 e 10
X = np.random.rand(20) * 10

# gera valores de Y com base em um coeficiente angular (b1), escalar (b0) e erro (epsilon)
b1 = 2.5
b0 = 5
epsilon = np.random.randn(20) * 2
Y = b1 + b0 * X + epsilon

# Método dos mínimos quadrados
def minimos_quadrados(X, Y):
    ux = X.mean()
    uy = Y.mean()

    # estima coeficiente angular
    B1 = np.sum((X - ux) * (Y - uy)) / np.sum((X - ux)**2)
    # estima coeficiente escalar
    B0 = uy - B1 * ux
    
    return B0, B1

# obtém a equação da linha de regressão usan do o método dos mínimos quadrados
B0_est, B1_est = minimos_quadrados(X, Y)
Y_est = B0_est + B1_est * X

# Visualização
plt.scatter(X, Y, label="Dados")
plt.plot(X, Y_est, color="red", label="Reta estimada")
for xi, yi, y_est in zip(X, Y, Y_est):
    plt.plot([xi, xi], [yi, y_est], color="gray", linestyle="-", linewidth=1)
plt.legend()
plt.show()


```
