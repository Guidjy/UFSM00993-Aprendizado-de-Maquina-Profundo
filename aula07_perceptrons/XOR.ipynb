```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0, x)

def i(x):
    return (x > 0).astype(int)


# entrada e saída da operação XOR
x = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 0])

# camada oculta
h_peso = np.array([[1, 1],
               [1, 1]])
h_vies = np.array([0, -1])
h = relu(x @ h_peso + h_vies)

# camada de saída
yhat_peso = np.array([1, -2])
yhat_vies = 0
yhat = i(h @ yhat_peso + yhat_vies)

# plota
colors = ['blue' if yi == 0 else 'red' for yi in yhat]
plt.scatter(x[:,0], x[:,1], c=colors, s=200)
plt.title("Função XOR")
plt.show()



```
